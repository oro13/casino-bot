```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
from scipy.stats import geom
```

model separately the come out and the point on

```python
#point on : (not modeled here)
n = 10 #trials

for i in range(n):
    report = []
    dice_sum = 0
    while dice_sum != 7:
        dice1 = np.random.randint(1,7)
        dice2 = np.random.randint(1,7)
        report.append(dice1+dice2)
        dice_sum = dice1+dice2
    print(f'\nRound finished in {len(report)} rolls:\n{report}')
```

prob of rolling a seven is 6/36



Let's explore the probability of going k trials before seeing a 7.

```python

p = 6/36
mean, var, skew, kurt = geom.stats(p, moments='mvsk')
```

```python

fig, ax = plt.subplots(1, 1, figsize=(14,8))

x = np.arange(geom.ppf(0.01, p),

              geom.ppf(0.99, p))

ax.plot(x, geom.pmf(x, p), 'bo', ms=8, label='geom pmf')

ax.vlines(x, 0, geom.pmf(x, p), colors='b', lw=3, alpha=0.5)

ax.legend(loc='best', frameon=False)
ax.set_xlabel('k', fontsize=18)
ax.set_ylabel('Probability', fontsize=18)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(14,8))
rv = geom(p)

ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,

        label='frozen pmf')

ax.legend(loc='best', frameon=False)
ax.set_xlabel('k', fontsize=18)
ax.set_ylabel('Probability', fontsize=18)
plt.tight_layout()
plt.show()
```
